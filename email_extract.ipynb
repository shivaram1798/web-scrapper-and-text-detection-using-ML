```python
import cv2
import pytesseract
import os
```

```python
pytesseract.pytesseract.tesseract_cmd='C:\\Program Files (x86)\\Tesseract-OCR\\tesseract.exe'
directory = 'D:\\freelancing\\insurance_pan_india'
die = 'res'
path = os.path.join(directory,die)
#os.mkdir(path)
```

```python
i=1
#for filename in os.listdir(directory):
for f in range(i,2):
    r=directory+"\\res\\"+str(i)+".txt"
    i=i+1
    res = open(r,"w")
    #f = os.path.join(directory, filename)
    fa = directory+"\\"+str(f)+".png"
    #print(fa)
    img= cv2.imread(fa)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    #cv2.imshow('result', img)
    res.write(pytesseract.image_to_string(img,lang="font_name"))
    #res.write(pytesseract.image_to_string(img))
    res.close()
    print(str(i-1)+" is done")
    #cv2.waitKey(0)
```
